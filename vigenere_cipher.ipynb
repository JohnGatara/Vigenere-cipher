Vigenere Cipher is a simple polyaphabetic substitution text encryption method.A polyalphabetic cipher is any cipher based on substitution using multiple offset alphabets. The source text is encrypted using the Vigen√®re table. The table consists of alphabets (array of characters) written out 26 times in different rows (for English alphabet), each alphabet is cyclically shifted to the left compared to the previous alphabet.

The encryption is done as follows:
Ci = (P i+ Kj) mod 26,
where Kj is the j-th letter of the key, 
Pi is —ñ-th letter of the original message, and
Ci is the i-th letter of the ciphertext.
This solution removes special caharacters and numbers from the message to be encrypted as it was done traditionally.


```python
def vigenere_decrypt(cipher, key):
  cipher =text_processor(cipher)
  x=0
  decrypted_text =""
  while x<len(cipher):
    d =(ord(cipher[x])-ord(key[x])+26)%26
    d +=ord('A')
    decrypted_text +=chr(d)
    x=x+1
  return decrypted_text
```

```python
def key_generator(plain, key):
  plain_len =len(plain)
  key_len = len(key)
  key = key.upper()
  i=0
  while i <= plain_len:  
    if plain_len == len(key):
      break
    key+=key[i]
    i =i+1

  return key
```

```python
def vigenere_encrypt(plain, key):
  x=0
  hashed_text =""
  while x<=len(plain)-1:
    c = (ord(plain[x])+ord(key[x]))%26
    c +=ord('A')
    hashed_text +=chr(c)
    x=x+1
  return hashed_text
```

```python
def text_processor(plain):
   new_plain = ''.join(char for char in plain if char.isalnum())  #remove special chracters
   nplain = new_plain.upper()
   nplain = ''.join([i for i in nplain if not i.isdigit()]) #remove numbers
   return nplain 
```
